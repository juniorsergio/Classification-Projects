Uma importante verificação nos dados é garantir que não existam valores nulos no dataset, que trazem inúmeros problemas ao criar os modelos de previsão. Abaixo, foram encontradas duas variáveis em que há ocorrência de valores nulos: 'salario_mensal' e 'numero_de_dependentes'.

# Modelagem

### Escolha dos modelos

Como descrito anteriormente, o dataset possui dois graves problemas que irão interferir na previsão de inadimplência: um desbalanceamento de classes (apenas 6,6% do dataset é de inadimplentes) e uma alta quantidade de valores nulos na variável 'salario_mensal' (19,8%).

Para o primeiro problema serão usadas duas abordagens: oversampling e ajuste de parâmetros compensatórios no modelo. A segunda abordagem será possível graças ao algoritmo XGBoost, que permite esse tipo de otimização sem que seja necessário alterar o dataset a priori. Isso é feito a partir de uma super compensação dos erros na classe positiva, o que pode gerar um overfitting se não for feito cuidadosamente. Como esse algortimo funciona a base de árvores de decisão, o Random Forest será utilizado como baseline.

Para o segundo problema, os dados nulos serão preenchidos usando o algortimo KNNImputer (explicado a seguir).

Assim, serão gerados cinco modelos usando dois algoritmos:

- Random Forest (default e com oversampling);
- XGBoost (default, com oversampling e otimizado para classes desbalanceadas).

### Preenchendo os valores nulos com KNNImputer

Considerando a volumetria de dados nulos na coluna 'salario_medio' (quase 20%), uma abordagem tradicional como o uso de média, mediana ou moda para preencher esses valores pode gerar um overfitting no modelo. Assim, será utilizado o método KNNImputer para estimar os valores nulos de acordo com o valor médio dos n=3 vizinhos. Como esse algortimo analisa o dataset como um todo, as ocorrências nulas em 'numero_de_dependentes' também passarão pelo mesmo tratamento.

<b>OBS:</b> o valor de n foi escolhido vizando otimizar o tempo de execução do algoritmo. 

```python
x = train.drop(columns=['inadimplente'])

scaler = MinMaxScaler()
x_trans = pd.DataFrame(scaler.fit_transform(x), columns = x.columns)
x_trans.head()
```

```python
imputer = KNNImputer(n_neighbors=3)
x_trans = pd.DataFrame(imputer.fit_transform(x_trans), columns=x_trans.columns)
```

```python
print(f"Proporção de valores nulos da variável salario_mensal:\n{x_trans['salario_mensal'].isna().value_counts('normalized')}")
print()
print(f"Proporção de valores nulos da variável numero_de_dependentes:\n{x_trans['numero_de_dependentes'].isna().value_counts('normalized')}")
```

### Divisão de treino e teste para os modelos

Inicialmente usou-se uma divisão 70/30 para os dados, já que os dados de teste reais possuem aproximadamente essa proporção com os dados de treino. Porém, ao rodar a divisão 60/40, viu-se que os resultados dos modelos foram ligeiramente melhores, o que justificou a escolha por essa divisão dos dados.

```python
x = pd.DataFrame(scaler.inverse_transform(x_trans), columns = x_trans.columns)
y = train['inadimplente']

x.head()
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, train_size=0.6, random_state=42)
```

### Balanceando as classes com Borderline SMOTE

O algortimo SMOTE utiliza a distância de uma ocorrência da classe minoritária com os seus vizinhos para gerar dados sintéticos. Entretanto, ele não consegue identificar quando uma entrada foge do padrão das demais, e acaba replicando-a e gerando ruído no modelo. Assim, será utilizado o Borderline SMOTE que consegue ignorar ruídos nos dados e gerar informações sintéticas considerando os pontos de fronteira entre as duas classes.

Fonte: https://towardsdatascience.com/class-imbalance-smote-borderline-smote-adasyn-6e36c78d804

```python
bd_smote = BorderlineSMOTE()
x_train_smote, y_train_smote = bd_smote.fit_resample(x_train, y_train)
sns.countplot(x=y_train_smote)
```

### Random Forest

```python
forest = RandomForestClassifier()
forest.fit(x_train, y_train)
```

```python
forest_pred = forest.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, forest_pred), annot=True, fmt='d')
print(classification_report(y_test, forest_pred))
```

### Random Forest com classes balanceadas

```python
forest_smote = RandomForestClassifier()
forest_smote.fit(x_train_smote, y_train_smote)
```

```python
forest_smote_pred = forest_smote.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, forest_smote_pred), annot=True, fmt='d')
print(classification_report(y_test, forest_smote_pred))
```

### XGBoost

```python
xgboost = XGBClassifier(use_label_encoder=False,
                        eval_metric='logloss')
xgboost.fit(x_train, y_train)
```

```python
xgboost_pred = xgboost.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, xgboost_pred), annot=True, fmt='d')
print(classification_report(y_test, xgboost_pred))
```

### XGBoost com classes balanceadas

```python
xgboost_smote = XGBClassifier(use_label_encoder=False,
                        eval_metric='logloss')
xgboost_smote.fit(x_train_smote, y_train_smote)
```

```python
xgboost_smote_pred = xgboost_smote.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, xgboost_smote_pred), annot=True, fmt='d')
print(classification_report(y_test, xgboost_smote_pred))
```

### XGBoost otimizado para classes desbalanceadas

A proporção recomendada (e utilizada nesse case) pela documentação do XGBoost para o parâmetro responsável por aumentar o peso dos erros na classe positiva é:
- scale_pos_weight = total_negative_examples / total_positive_examples

Também foi adicionado o parâmetro max_delta_step, responsável por atribuir um peso máximo a cada nó-folha. Isso é necessário em casos de classes desbalanceadas pois o peso de cada nó/árvore tende para a classe majoritária, o que prejudica a assertividade na classe que realmente importa (no caso, detectar inadimplentes).

```python
scale_pos_weight = y_train.value_counts()[0] / y_train.value_counts()[1]
print(f"Escala para rebalanceamento: {scale_pos_weight}")
```

```python
xgboost_balanced = XGBClassifier(scale_pos_weight=scale_pos_weight,
                                 use_label_encoder=False,
                                 eval_metric='logloss', # padrão para classificação
                                 max_delta_step=1)
xgboost_balanced.fit(x_train, y_train)
```

```python
xgboost_balanced_pred = xgboost_balanced.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, xgboost_balanced_pred), annot=True, fmt='d')
print(classification_report(y_test, xgboost_balanced_pred))
```

### Conclusão

Considerando o problema apresentado de prever se um indivíduo se tornará ou não inadimplente no futuro, o cenário mais interessante para uma empresa é ser capaz de capturar o maior número possível de casos positivos, evitando prejuízos futuros.

Assim, dentro os cinco modelos gerados, o único que conseguiu atender esse propósito foi o XGBoost parametrizado para considerar o desbalanceamento de classes no dataset. A justificativa é que, <b>de todos os casos que foram apontados como não inadimplentes no dataset de teste, o modelo foi capaz de acertar 98%.</b> Ou seja, a margem de erro de 2% torna seguro aprovar automaticamente aqueles que assim forem classificados pelo modelo. Em contra partida, <b>de todos os indivíduos que de fato se tornariam inadimplentes, o modelo foi capaz de capturar 70% deles.</b> Infelizmente há um trade-off nessa operaçã, já que a precisão do modelo para essa classe é baixa (23%). Entretanto, no escopo desse problema, vale a pena para a empresa ter um modelo que barre mais pessoas (mesmo que com baixa acertividade) para que uma análise mais cuidadosa por um agente humano seja feita e menos pessoas propensas a inadimplência tenham o seu crédito aprovado diretamente.

Apesar do modelo obtido ser satisfatório para o objetivo proposto, demais investigações e tratamentos podem e devem ser feitos para se obter um melhor equilíbrio entre a precisão e o recall da classe de inadimplentes. Uma alternativa seria usar undersampling na classe majoritária ao invés do oversampling utilizado nesse case. Técnicas como GridSearch também poderiam ser utilizadas para parametrizar melhor os algoritmos escolhidos. Além disso, seria interessante também rodar algoritmos de seleção de features para analisar se as variáveis independentes informadas pelo negócio são de fato relevantes para a geração de um bom modelo.
