# Modelagem

### Escolha dos modelos

Como descrito anteriormente, o dataset possui dois graves problemas que irÃ£o interferir na previsÃ£o de inadimplÃªncia: um desbalanceamento de classes (apenas 6,6% do dataset Ã© de inadimplentes) e uma alta quantidade de valores nulos na variÃ¡vel 'salario_mensal' (19,8%).

Para o primeiro problema serÃ£o usadas duas abordagens: oversampling e ajuste de parÃ¢metros compensatÃ³rios no modelo. A segunda abordagem serÃ¡ possÃ­vel graÃ§as ao algoritmo XGBoost, que permite esse tipo de otimizaÃ§Ã£o sem que seja necessÃ¡rio alterar o dataset a priori. Isso Ã© feito a partir de uma super compensaÃ§Ã£o dos erros na classe positiva, o que pode gerar um overfitting se nÃ£o for feito cuidadosamente. Como esse algortimo funciona a base de Ã¡rvores de decisÃ£o, o Random Forest serÃ¡ utilizado como baseline.

Para o segundo problema, os dados nulos serÃ£o preenchidos usando o algortimo KNNImputer (explicado a seguir).

Assim, serÃ£o gerados cinco modelos usando dois algoritmos:

- Random Forest (default e com oversampling);
- XGBoost (default, com oversampling e otimizado para classes desbalanceadas).

### Preenchendo os valores nulos com KNNImputer

Considerando a volumetria de dados nulos na coluna 'salario_medio' (quase 20%), uma abordagem tradicional como o uso de mÃ©dia, mediana ou moda para preencher esses valores pode gerar um overfitting no modelo. Assim, serÃ¡ utilizado o mÃ©todo KNNImputer para estimar os valores nulos de acordo com o valor mÃ©dio dos n=3 vizinhos. Como esse algortimo analisa o dataset como um todo, as ocorrÃªncias nulas em 'numero_de_dependentes' tambÃ©m passarÃ£o pelo mesmo tratamento.

<b>OBS:</b> o valor de n foi escolhido vizando otimizar o tempo de execuÃ§Ã£o do algoritmo. 

```python
x = train.drop(columns=['inadimplente'])

scaler = MinMaxScaler()
x_trans = pd.DataFrame(scaler.fit_transform(x), columns = x.columns)
x_trans.head()
```

```python
imputer = KNNImputer(n_neighbors=3)
x_trans = pd.DataFrame(imputer.fit_transform(x_trans), columns=x_trans.columns)
```

```python
print(f"ProporÃ§Ã£o de valores nulos da variÃ¡vel salario_mensal:\n{x_trans['salario_mensal'].isna().value_counts('normalized')}")
print()
print(f"ProporÃ§Ã£o de valores nulos da variÃ¡vel numero_de_dependentes:\n{x_trans['numero_de_dependentes'].isna().value_counts('normalized')}")
```

### DivisÃ£o de treino e teste para os modelos

Inicialmente usou-se uma divisÃ£o 70/30 para os dados, jÃ¡ que os dados de teste reais possuem aproximadamente essa proporÃ§Ã£o com os dados de treino. PorÃ©m, ao rodar a divisÃ£o 60/40, viu-se que os resultados dos modelos foram ligeiramente melhores, o que justificou a escolha por essa divisÃ£o dos dados.

```python
x = pd.DataFrame(scaler.inverse_transform(x_trans), columns = x_trans.columns)
y = train['inadimplente']

x.head()
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, train_size=0.6, random_state=42)
```

### Balanceando as classes com Borderline SMOTE

O algortimo SMOTE utiliza a distÃ¢ncia de uma ocorrÃªncia da classe minoritÃ¡ria com os seus vizinhos para gerar dados sintÃ©ticos. Entretanto, ele nÃ£o consegue identificar quando uma entrada foge do padrÃ£o das demais, e acaba replicando-a e gerando ruÃ­do no modelo. Assim, serÃ¡ utilizado o Borderline SMOTE que consegue ignorar ruÃ­dos nos dados e gerar informaÃ§Ãµes sintÃ©ticas considerando os pontos de fronteira entre as duas classes.

Fonte: https://towardsdatascience.com/class-imbalance-smote-borderline-smote-adasyn-6e36c78d804

```python
bd_smote = BorderlineSMOTE()
x_train_smote, y_train_smote = bd_smote.fit_resample(x_train, y_train)
sns.countplot(x=y_train_smote)
```

### Random Forest

```python
forest = RandomForestClassifier()
forest.fit(x_train, y_train)
```

```python
forest_pred = forest.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, forest_pred), annot=True, fmt='d')
print(classification_report(y_test, forest_pred))
```

### Random Forest com classes balanceadas

```python
forest_smote = RandomForestClassifier()
forest_smote.fit(x_train_smote, y_train_smote)
```

```python
forest_smote_pred = forest_smote.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, forest_smote_pred), annot=True, fmt='d')
print(classification_report(y_test, forest_smote_pred))
```

### XGBoost

```python
xgboost = XGBClassifier(use_label_encoder=False,
                        eval_metric='logloss')
xgboost.fit(x_train, y_train)
```

```python
xgboost_pred = xgboost.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, xgboost_pred), annot=True, fmt='d')
print(classification_report(y_test, xgboost_pred))
```

### XGBoost com classes balanceadas

```python
xgboost_smote = XGBClassifier(use_label_encoder=False,
                        eval_metric='logloss')
xgboost_smote.fit(x_train_smote, y_train_smote)
```

```python
xgboost_smote_pred = xgboost_smote.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, xgboost_smote_pred), annot=True, fmt='d')
print(classification_report(y_test, xgboost_smote_pred))
```

### XGBoost otimizado para classes desbalanceadas

A proporÃ§Ã£o recomendada (e utilizada nesse case) pela documentaÃ§Ã£o do XGBoost para o parÃ¢metro responsÃ¡vel por aumentar o peso dos erros na classe positiva Ã©:
- scale_pos_weight = total_negative_examples / total_positive_examples

TambÃ©m foi adicionado o parÃ¢metro max_delta_step, responsÃ¡vel por atribuir um peso mÃ¡ximo a cada nÃ³-folha. Isso Ã© necessÃ¡rio em casos de classes desbalanceadas pois o peso de cada nÃ³/Ã¡rvore tende para a classe majoritÃ¡ria, o que prejudica a assertividade na classe que realmente importa (no caso, detectar inadimplentes).

```python
scale_pos_weight = y_train.value_counts()[0] / y_train.value_counts()[1]
print(f"Escala para rebalanceamento: {scale_pos_weight}")
```

```python
xgboost_balanced = XGBClassifier(scale_pos_weight=scale_pos_weight,
                                 use_label_encoder=False,
                                 eval_metric='logloss', # padrÃ£o para classificaÃ§Ã£o
                                 max_delta_step=1)
xgboost_balanced.fit(x_train, y_train)
```

```python
xgboost_balanced_pred = xgboost_balanced.predict(x_test)
```

```python
sns.heatmap(confusion_matrix(y_test, xgboost_balanced_pred), annot=True, fmt='d')
print(classification_report(y_test, xgboost_balanced_pred))
```

### ConclusÃ£o

Considerando o problema apresentado de prever se um indivÃ­duo se tornarÃ¡ ou nÃ£o inadimplente no futuro, o cenÃ¡rio mais interessante para uma empresa Ã© ser capaz de capturar o maior nÃºmero possÃ­vel de casos positivos, evitando prejuÃ­zos futuros.

Assim, dentro os cinco modelos gerados, o Ãºnico que conseguiu atender esse propÃ³sito foi o XGBoost parametrizado para considerar o desbalanceamento de classes no dataset. A justificativa Ã© que, <b>de todos os casos que foram apontados como nÃ£o inadimplentes no dataset de teste, o modelo foi capaz de acertar 98%.</b> Ou seja, a margem de erro de 2% torna seguro aprovar automaticamente aqueles que assim forem classificados pelo modelo. Em contra partida, <b>de todos os indivÃ­duos que de fato se tornariam inadimplentes, o modelo foi capaz de capturar 70% deles.</b> Infelizmente hÃ¡ um trade-off nessa operaÃ§Ã£, jÃ¡ que a precisÃ£o do modelo para essa classe Ã© baixa (23%). Entretanto, no escopo desse problema, vale a pena para a empresa ter um modelo que barre mais pessoas (mesmo que com baixa acertividade) para que uma anÃ¡lise mais cuidadosa por um agente humano seja feita e menos pessoas propensas a inadimplÃªncia tenham o seu crÃ©dito aprovado diretamente.

Apesar do modelo obtido ser satisfatÃ³rio para o objetivo proposto, demais investigaÃ§Ãµes e tratamentos podem e devem ser feitos para se obter um melhor equilÃ­brio entre a precisÃ£o e o recall da classe de inadimplentes. Uma alternativa seria usar undersampling na classe majoritÃ¡ria ao invÃ©s do oversampling utilizado nesse case. TÃ©cnicas como GridSearch tambÃ©m poderiam ser utilizadas para parametrizar melhor os algoritmos escolhidos. AlÃ©m disso, seria interessante tambÃ©m rodar algoritmos de seleÃ§Ã£o de features para analisar se as variÃ¡veis independentes informadas pelo negÃ³cio sÃ£o de fato relevantes para a geraÃ§Ã£o de um bom modelo.
